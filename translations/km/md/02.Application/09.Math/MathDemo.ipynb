## бһ“бһ¶бҹҶбһ…бһјбһӣбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢ
бһ“бһ¶бҹҶбһ…бһјбһӣбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ PyTorch бһ“бһ·бһ„ transformers бһҠбҹӮбһӣбһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ•бҹ’бһ‘бһ»бһҖ бһ“бһ·бһ„бһ”бҹ’бһҡбһҫбһҳбҹүбһјбһҠбҹӮбһӣ Phi-4бҹ”


```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

## бһҖбҹҶбһҺбһҸбҹӢбһҹбҹҠбһёбһҠбһ…бҹғбһҠбһ“бҹ’бһҷ
бһҖбҹҶбһҺбһҸбҹӢбһҹбҹҠбһёбһҠбһ…бҹғбһҠбһ“бҹ’бһҷ бһҠбһҫбһҳбҹ’бһ”бһёбһ’бһ¶бһ“бһ¶бһҗбһ¶бһӣбһ‘бҹ’бһ’бһ•бһӣбһўбһ¶бһ…бһҹбҹ’бһҸбһ¶бһҡбһЎбһҫбһ„бһңбһ·бһүбһ”бһ¶бһ“бһ“бҹ…бһ–бҹҒбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ•бҹ’бһҹбҹҒбһ„бҹ—


```python
torch.random.manual_seed(0)
```

## бһ•бҹ’бһ‘бһ»бһҖ Phi-4-mini-flash-reasoning Model бһ“бһ·бһ„ Tokenizer
бһ•бҹ’бһ‘бһ»бһҖ Microsoft Phi-4-mini-flash-reasoningmodel бһ“бһ·бһ„ tokenizer бһҠбҹӮбһӣбһ–бһ¶бһҖбҹӢбһ–бҹҗбһ“бҹ’бһ’бһ–бһё Hugging Faceбҹ” бһҳбҹүбһјбһҠбҹӮбһӣбһ“бҹҒбҹҮбһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ•бҹ’бһ‘бһ»бһҖбһӣбһҫ CUDA бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫ inference бһ”бһ¶бһ“бһӣбһҝбһ“бҹ”


```python
model_id = "microsoft/Phi-4-mini-flash-reasoning"
model = AutoModelForCausalLM.from_pretrained(
   model_id,
   device_map="cuda",
   torch_dtype="auto",
   trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

## бһҡбҹҖбһ”бһ…бҹҶбһҹбһ¶бһҡвҖӢбһ”бһүбҹ’бһ…бһјбһӣ
бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбһ¶бһҡвҖӢбһҹбһ“бҹ’бһ‘бһ“бһ¶вҖӢбһҳбһҪбһҷвҖӢбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бһүбҹ’бһ бһ¶бһӮбһҺбһ·бһҸбһңбһ·бһ‘бҹ’бһҷбһ¶вҖӢбһҮбһ¶вҖӢбһҹбһҳбһёбһҖбһ¶бһҡбһҗбҹ’бһ“бһ¶бһҖбҹӢбһ‘бһёбҹў бһ бһҫбһҷбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһңбһ¶вҖӢбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһӮбҹҶбһҡбһјбһҹбһ“бҹ’бһ‘бһ“бһ¶бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбҹүбһјбһҢбҹӮбһӣбҹ”


```python
messages = [{
   "role": "user",
   "content": "How to solve 3*x^2+4*x+5=1?"
}]   
inputs = tokenizer.apply_chat_template(
   messages,
   add_generation_prompt=True,
   return_dict=True,
   return_tensors="pt",
)
```

## бһ”бһ„бҹ’бһҖбһҫбһҸбһ…бһҳбҹ’бһӣбһҫбһҷ
бһ”бһ„бҹ’бһҖбһҫбһҸбһ…бһҳбҹ’бһӣбһҫбһҷбһ–бһёбһҳбҹүбһјбһҢбҹӮбһӣбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢ бһҠбһјбһ…бһҮбһ¶ temperature бһ“бһ·бһ„ top_p бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ—бһ¶бһ–бһ…бҹғбһҠбһ“бҹ’бһҷбһҖбҹ’бһ“бһ»бһ„бһӣбһ‘бҹ’бһ’бһ•бһӣбҹ”


```python
outputs = model.generate(
   **inputs.to(model.device),
   max_new_tokens=32768,
   temperature=0.6,
   top_p=0.95,
   do_sample=True,
)
```

## бһ”бһҖбһҹбҹ’бһҡбһ¶бһҷбһӣбһ‘бҹ’бһ’бһ•бһӣбһ‘бҹ…бһҮбһ¶бһўбһҸбҹ’бһҗбһ”бһ‘

бһ”бһҳбҹ’бһӣбҹӮбһ„бһӣбҹҶбһҠбһ¶бһ”бҹӢбһҸбһјбһҖбһ·бһ“бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһҸбҹ’бһҡбһЎбһ”бҹӢбһ‘бҹ…бһҮбһ¶бһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹӮбһӣбһҳбһ“бһ»бһҹбҹ’бһҹбһўбһ¶бһ…бһўбһ¶бһ“бһ”бһ¶бһ“, бһҠбҹ„бһҷбһҳбһ·бһ“бһҡбһ¶бһ”бҹӢбһ”бһүбҹ’бһ…бһјбһӣбһҸбһјбһҖбһ·бһ“бһ“бҹғбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҠбһҫбһҳ бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһ бһ¶бһүбһҸбҹӮбһ…бһҳбҹ’бһӣбһҫбһҷбһҡбһ”бһҹбҹӢбһҳбҹүбһјбһҠбҹӮбһӣбҹ”


```python
outputs = tokenizer.batch_decode(outputs[:, inputs["input_ids"].shape[-1]:])
```
